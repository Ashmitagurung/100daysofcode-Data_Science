**Zomato Dataset Exploratory Data Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/zomato.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
## Missing Values
df.isnull().sum()
```

```python
# Another way of find missing values
#Use of list comprehension
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df.shape  # means 9551 rows so we cannot see through heatmap
```

```python
# Country code
```

```python
df_country = pd.read_excel('/content/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
df_country.shape
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val = final_df.Country.value_counts().values
```

```python
country_val
```

```python
# pie chart

plt.pie(country_val, labels=country_names, autopct='%1.2f%%')
plt.show()
```

```python
# Top 3 countrys that  uses zomato

plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

```python
# Explore numerical variables
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})

```

```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)


```

```python
#count plot
sns.countplot(x='Rating color', data=ratings, palette=['orange','blue', 'red', 'yellow','green',])
```
