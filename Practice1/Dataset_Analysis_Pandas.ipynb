# Handling Missing Values

```python
weather = pd.read_csv('/content/weather_data1.csv')
```

```python
weather
```

```python
# Identyfying missing values
weather.isnull().sum()
```

```python
weather.notnull().sum()
```

```python
weather.fillna(0)
```

```python
weather.fillna({"temperature":0,"windspeed":0,"event":"No Event"})
```

```python
weather.fillna({"temperature":weather.temperature.mean(),"windspeed":weather.windspeed.mean(),"event":"No Event"})
```

```python

weather.fillna(method="ffill")
```

```python

weather.fillna(method="bfill")
```

```python
weather
```
