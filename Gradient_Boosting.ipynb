Gradient Boosting

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Titanic.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns = ['sex', 'embarked', 'class', 'who'])
```

```python
df
```

```python
df.info()
```

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
```

```python
df
```

```python
df.drop(columns=['sibsp', 'parch'], inplace=True)
```

```python
df
```

```python
X=df.drop(columns=['survived'])
y=df['survived']
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier,AdaBoostClassifier
from sklearn.model_selection import GridSearchCV,KFold,cross_val_score
```

```python
ra = RandomForestClassifier(random_state = 1)
```

```python
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
```

```python
kf
```

```python
cross_val_results = cross_val_score(ra, X, y, cv=kf,verbose = 5)
```

```python
cross_val_results.mean()
```

```python
ra.fit(X_train, y_train)
```

```python
ra.get_params()
```

```python
parameters = {'n_estimators':(10,50,100,1000),
              'criterion':('gini', 'entropy', 'log_loss'),
              'max_features':('sqrt','log2')
              }
grid=GridSearchCV(estimator=ra,param_grid=parameters,cv=5,verbose = 5)
```

```python
grid.fit(X,y)
```

```python
grid.best_score_
```

```python
grid.best_score_
```

```python
pd.DataFrame(grid.cv_results_)
```

```python
grid.best_estimator_
```

```python
cl = GradientBoostingClassifier(random_state=1)
```

```python
cl.fit(X_train,y_train)
```

```python
cl.score(X_train, y_train)
```

```python
cl.score(X_test, y_test)
```

```python
ra.score(X_train,y_train)
```

```python
ra.score(X_test,y_test)
```

```python
ad = AdaBoostClassifier(random_state=1)
```

```python
ad.fit(X_train, y_train)
```

```python
ad.score(X_train, y_train)
```

```python
ad.score(X_test, y_test)
```

```python
ad2 = AdaBoostClassifier(random_state=100,n_estimators=10,learning_rate=0.1,base_estimator=RandomForestClassifier(random_state=1,n_estimators=10))
```

```python
ad2.fit(X_train,y_train)
```

```python
ad2.score(X_train,y_train)
```
