```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.arange(0,10)*np.random.randint(0, 10, 10)

plt.plot(y, marker= 'v')
plt.show()
```

```python
x = np.array([0, 4, 5, 1, 6, 7])
y = np.array([0, 1, 5, 1, 2, 7])

plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title(label = "Sample Graph")

plt.plot(x,y, marker = 'o', color = 'red', label = 'line 1', linewidth = 2)
plt.plot(y, marker = 'o', color = 'blue', label = 'line 2')
plt.plot()
plt.legend()
plt.grid()

plt.show()
```

```python
x=np.array([0,2,3,7,9,12])
y=np.array([0,4,2,6,5,10])
plt.figure(figsize = (20,20), edgecolor = "red")
plt.plot(x, y, marker = '.', label = "ABC")
plt.plot(x+1, y+2, marker = 'v', label = 'BAC')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Stock Prize')
plt.legend()
plt.grid()
plt.show()
```

```python
x= np.random.normal(5,5,1000)
y=np.random.normal(10,2,1000)

col=(x+y)/2
plt.scatter(x,y,c=col,cmap='Blues')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title("different daigrams")

plt.subplot(1,3,3)
plt.scatter(x,y)
plt.title('Scatter plot')

plt.subplot(1,3,2)
plt.hist(x)
plt.title('Hist plot of x')

plt.subplot(1,3,1)
plt.hist(y,orientation='horizontal')
plt.title('Hist plot of y')
```
