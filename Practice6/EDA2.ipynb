**Black Friday Dataset EDA and Feature Engineering**

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing the dataset
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test
```

```python
##MErge both train and test data
df = pd.concat([df_train, df_test])
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)

#axis = 0 -> horizontally
# axis = 1 -> verically
```

```python
df.head()
```

```python
#Handing Categorical feature gender
df['Gender'] = pd.get_dummies(df['Gender'], drop_first=1)
```

```python
df['Gender']
```

```python
# Handling categorical feature age

df['Age'].unique()
```

```python
# pd.get_dummies(df['Age'], drop_first=True)
```

```python
df['Age'] = df['Age'].map({'0-17': 1, '18-25': 2, '26-35': 3, '36-45':4, '46-50': 5, '51-55': 6, '55+':7})
```

```python
df['Age']
```

```python
# Another techniques to convert the categorical value into the numeric value
# Most useful technique

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df['Age'] = label_encoder.fit_transform(df['Age'])

df['Age'].unique()
```

```python
# fixing categorical City_category
df_city = pd.get_dummies(df['City_Category'], drop_first=1)
```

```python
df_city
```

```python
df = pd.concat([df, df_city], axis=1)
```

```python

df.head()
```

```python
df.drop('City_Category', axis = 1)
```

```python
df.head()
```

```python
#Check for missing values

df.isnull().sum()
```

```python
# Focus on replacing missing values

df['Product_Category_2'].unique()
```

```python
# Replace missing vlaues

df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode())
```

```python

df['Product_Category_2']
```

```python
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode())
```
