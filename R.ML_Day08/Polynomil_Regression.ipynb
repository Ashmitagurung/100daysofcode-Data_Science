# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

```python
data_baseball = pd.read_csv('/content/drive/MyDrive/data/baseball.csv')
```

```python
data_baseball
```

```python
data_baseball.info()
```

```python
X = data_baseball.drop(columns = ['player', 'salary'])
y = data_baseball['salary']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```

```python
X_train.shape, X_test.shape
```

```python
model_lasso = Lasso(alpha = 0.05, max_iter = 10000, tol = 0.000001)
```

```python
model_lasso.fit(X_train, y_train)
```

```python
model_lasso.score(X_train, y_train)
```

```python
model_lasso.score(X_test, y_test)
```

```python
model_lasso.coef_
```

# Ridge

```python
from sklearn.linear_model import Ridge
```

```python
model_ridge = Ridge(alpha = 100, max_iter = 1000000, tol = 0.00000001)
```

```python
model_ridge.fit(X_train, y_train)
```

```python
model_ridge.score(X_train, y_train)
```

```python
model_ridge.score(X_test, y_test)
```

```python
model_ridge.coef_
```
